```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix, ConfusionMatrixDisplay

def plot_patient_confusion(csv_file=None, patient_id=None, models_folder="patient_models50_csv", class_names=None):
    
    # تحديد الملف تلقائيًا
    if csv_file is None and patient_id is not None:
        base_path = os.path.join(models_folder)
        
        # نحاول أكثر من صيغة ملف
        possible_files = [
            f"patient_{int(float(patient_id))}_model.csv",
            f"patient_{float(patient_id)}_model.csv",
            f"patient_{int(float(patient_id))}_validation.csv",
            f"patient_{float(patient_id)}_validation.csv",
            f"patient_{int(float(patient_id))}_test.csv",
            f"patient_{float(patient_id)}_test.csv",
        ]
        
        csv_file = None
        for file in possible_files:
            temp_path = os.path.join(base_path, file)
            if os.path.exists(temp_path):
                csv_file = temp_path
                break
        
        if csv_file is None:
            raise FileNotFoundError("No matching patient file found.")
    
    elif csv_file is None:
        raise ValueError("You must provide either csv_file or patient_id")

    # قراءة الملف
    df_patient = pd.read_csv(csv_file)

    # التأكد من الأعمدة
    if 'label' not in df_patient.columns or 'pred_label' not in df_patient.columns:
        raise ValueError("CSV must contain 'label' and 'pred_label' columns")

    y_true = df_patient['label'].values
    y_pred = df_patient['pred_label'].values

    # Accuracy
    accuracy = accuracy_score(y_true, y_pred)

    # Recall (Ictal class = 2)
    recall_ictal = recall_score(
        y_true,
        y_pred,
        labels=[2],
        average='micro',
        zero_division=0
    )

    print(f"\nPatient Metrics for file: {csv_file}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Recall (Ictal class): {recall_ictal:.4f}")

    # أسماء الكلاسات
    if class_names is None:
        class_names = ['Ictal', 'Normal', 'Pre-ictal']

    # Confusion Matrix Normalized
    cm_norm = confusion_matrix(y_true, y_pred, normalize='true')
    plt.figure(figsize=(7, 7))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm_norm, display_labels=class_names)
    disp.plot(cmap='Blues', values_format='.2f')
    plt.title("Normalized Confusion Matrix")
    plt.show()

    # Confusion Matrix Raw
    cm_raw = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(7, 7))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm_raw, display_labels=class_names)
    disp.plot(values_format='d')
    plt.title("Raw Confusion Matrix")
    plt.show()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    precision_recall_fscore_support,
    roc_auc_score
)
from sklearn.preprocessing import label_binarize
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, log_loss

def evaluate_all_patients(models_folder="patient_models50_csv", class_names=None):

    if class_names is None:
        class_names = ['Ictal', 'Normal', 'Pre-ictal']

    patient_files = sorted([
        f for f in os.listdir(models_folder)
        if f.endswith(".csv")
    ])

    all_y_true = []
    all_y_pred = []

    patient_acc = []
    patient_loss = []
    patient_ids = []

    for file in patient_files:

        file_path = os.path.join(models_folder, file)
        df = pd.read_csv(file_path)

        # التأكد من الأعمدة
        if 'label' not in df.columns or 'pred_label' not in df.columns:
            continue

        y_true = df['label'].values
        y_pred = df['pred_label'].values

        # Accuracy
        acc = accuracy_score(y_true, y_pred)

        # Log Loss (لو فيه probabilities)
        if 'prob_0' in df.columns:
            y_proba = df[['prob_0', 'prob_1', 'prob_2']].values
            loss = log_loss(y_true, y_proba)
        else:
            loss = np.nan  # لو مش موجود probabilities

        patient_acc.append(acc)
        patient_loss.append(loss)

        # استخراج patient id من اسم الملف
        patient_ids.append(file)

        all_y_true.extend(y_true)
        all_y_pred.extend(y_pred)

    return (
        np.array(all_y_true),
        np.array(all_y_pred),
        patient_acc,
        patient_loss,
        patient_ids
    )
```

```python
y_true, y_pred, patient_acc, patient_loss, patient_ids = evaluate_all_patients()
```

```python
plt.figure(figsize=(10,5))
plt.plot(patient_acc, marker='o')
plt.title("Accuracy per Patient")
plt.xlabel("Patient Index")
plt.ylabel("Accuracy")
plt.grid()
plt.show()

print(f"\nAverage Accuracy: {np.mean(patient_acc):.4f}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    precision_recall_fscore_support,
    log_loss,
    precision_recall_curve
)
from sklearn.preprocessing import label_binarize


# ===============================
# 1. Evaluate Patients Function
# ===============================
def evaluate_all_patients(models_folder="patient_models50_csv", class_names=None):

    if class_names is None:
        class_names = ['Ictal', 'Normal', 'Pre-ictal']

    patient_files = sorted([
        f for f in os.listdir(models_folder)
        if f.endswith(".csv")
    ])

    all_y_true = []
    all_y_pred = []

    patient_acc = []
    patient_loss = []
    patient_ids = []

    all_proba = []

    for file in patient_files:

        file_path = os.path.join(models_folder, file)

        try:
            df = pd.read_csv(file_path, encoding='latin1')
        except Exception as e:
            print(f"Skipping {file}: {e}")
            continue

        if 'label' not in df.columns or 'pred_label' not in df.columns:
            continue

        y_true = df['label'].values
        y_pred = df['pred_label'].values

        acc = accuracy_score(y_true, y_pred)

        # Log loss if probabilities exist
        if {'prob_0','prob_1','prob_2'}.issubset(df.columns):
            y_proba = df[['prob_0','prob_1','prob_2']].values
            loss = log_loss(y_true, y_proba)
            all_proba.append(y_proba)
        else:
            loss = np.nan

        patient_acc.append(acc)
        patient_loss.append(loss)
        patient_ids.append(file)

        all_y_true.extend(y_true)
        all_y_pred.extend(y_pred)

    return (
        np.array(all_y_true),
        np.array(all_y_pred),
        patient_acc,
        patient_loss,
        patient_ids,
        np.vstack(all_proba) if len(all_proba) > 0 else None
    )


# ===============================
# 2. Run Evaluation
# ===============================
y_true, y_pred, patient_acc, patient_loss, patient_ids, y_proba_all = evaluate_all_patients()


# ===============================
# 3. Accuracy per patient
# ===============================
plt.figure(figsize=(10,5))
plt.plot(patient_acc, marker='o')
plt.title("Accuracy per Patient")
plt.xlabel("Patient Index")
plt.ylabel("Accuracy")
plt.grid()
plt.show()

print(f"\nAverage Accuracy: {np.mean(patient_acc):.4f}")


# ===============================
# 4. Confusion Matrix
# ===============================
class_names = ['Ictal', 'Normal', 'Pre-ictal']

cm = confusion_matrix(y_true, y_pred)
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

plt.figure(figsize=(6,5))
plt.imshow(cm_norm, cmap='Blues')
plt.title("Normalized Confusion Matrix")
plt.colorbar()

plt.xticks(range(3), class_names)
plt.yticks(range(3), class_names)

for i in range(3):
    for j in range(3):
        plt.text(j, i, f"{cm_norm[i,j]:.2f}",
                 ha='center', va='center')

plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()


# ===============================
# 5. Class-wise metrics
# ===============================
precision, recall, f1, _ = precision_recall_fscore_support(
    y_true, y_pred, labels=[0,1,2], zero_division=0
)

print("\nSeizure-Focused Metrics:")
print(f"Ictal Recall    : {recall[0]:.4f}")
print(f"Preictal Recall : {recall[2]:.4f}")
print(f"Preictal F1     : {f1[2]:.4f}")


# ===============================
# 6. Global metrics
# ===============================
macro_f1 = np.mean(f1)

y_true_bin = label_binarize(y_true, classes=[0,1,2])

roc_auc = np.nan
if y_proba_all is not None:
    roc_auc = roc_auc_score(
        y_true_bin,
        y_proba_all,
        multi_class='ovr',
        average='macro'
    )

print("\nGlobal Metrics:")
print(f"Macro F1-score : {macro_f1:.4f}")
print(f"Macro ROC-AUC  : {roc_auc:.4f}")


# ===============================
# 7. Precision-Recall Curve
# ===============================
if y_proba_all is not None:

    plt.figure(figsize=(8,6))

    y_true_bin = label_binarize(y_true, classes=[0,1,2])

    for i, class_name in enumerate(class_names):

        precision, recall, _ = precision_recall_curve(
            y_true_bin[:, i],
            y_proba_all[:, i]
        )

        plt.plot(recall, precision, label=class_name)

    plt.title("Precision-Recall Curve (Multi-Class)")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.legend()
    plt.grid()
    plt.show()
```
